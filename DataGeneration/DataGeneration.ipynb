```python
# This is a sample Python script.
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
```

```python
# Generate a synthetic dataset with easy separability
# X represents all the inputs to the model (e.g., age & height or blood pressure & cholesterol, etc.)
# X is also referred to as features or independent variables
# y represents the target variable (e.g., whether a person has a disease or not)
# y is also referred to as the label or dependent variable
X, y = make_classification(
    n_samples=10000,      # Total number of samples
    n_features=2,         # Number of features (dimensions)
    n_informative=1,     # Number of informative features
    n_redundant=0,       # Number of redundant features
    n_repeated=0,
    n_classes=2,
    n_clusters_per_class=1,
    flip_y=0,            # Fraction of samples whose class is randomly exchanged
    class_sep=5,         # Separation between the classes
    weights = [0.5, 0.5], # Can adjust balance of the data
    hypercube=True,
    random_state=42
)

print(X.shape)
print(y.shape)
```

```python
# Plot the generated dataset
plt.figure(figsize=(10, 6))
plt.scatter(X[:,0], X[:,1], c=y, cmap='coolwarm', marker='o', edgecolor='k', s=70, label='Data')
plt.title("Synthetic Data for ANN with Easy Separability")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

# Generate synthetic linear regression data
X, y = make_regression(n_samples=100, n_features=1, noise=15, random_state=42)

# Visualize the data
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', alpha=0.6, edgecolors='k')
plt.xlabel('Feature (X)')
plt.ylabel('Target (y)')
plt.title('Linear Regression Data')
plt.grid(True, alpha=0.3)
plt.show()

# Display data statistics
print(f"Number of samples: {X.shape[0]}")
print(f"Number of features: {X.shape[1]}")
print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
print(f"\nFirst 5 samples:")
print(f"X[:5]:\n{X[:5]}")
print(f"y[:5]: {y[:5]}")
```

```python
# Generate 1000 points on the line y=3x+2 with added noise
X_new = np.random.uniform(-2, 2, size=(1000, 1))
y_new = 3 * X_new.flatten() + 2 + np.random.normal(0, 1, size=1000)

# Visualize the generated data
plt.figure(figsize=(10, 6))
plt.scatter(X_new, y_new, color='green', alpha=0.5, s=30, edgecolors='k')
plt.plot(X_new, 3 * X_new + 2, color='red', linewidth=2, label='y=3x+2')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Generated Data: y=3x+2 with Noise')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"Generated {len(X_new)} points")
print(f"X_new shape: {X_new.shape}")
print(f"y_new shape: {y_new.shape}")
```

```python
from sklearn.linear_model import LinearRegression

# Perform linear regression
model = LinearRegression()
model.fit(X_new, y_new)

# Get the coefficients and intercept
slope = model.coef_[0]
intercept = model.intercept_

print(f"Linear Regression Results:")
print(f"Slope: {slope:.4f}")
print(f"Intercept: {intercept:.4f}")
print(f"Equation: y = {slope:.4f}x + {intercept:.4f}")
print(f"\nExpected equation: y = 3x + 2")
```

```python
# Visualize the linear regression results
plt.figure(figsize=(10, 6))
plt.scatter(X_new, y_new, color='green', alpha=0.5, s=30, edgecolors='k', label='Data')
plt.plot(X_new, model.predict(X_new), color='blue', linewidth=2, label=f'Fitted line: y={slope:.4f}x+{intercept:.4f}')
plt.plot(X_new, 3 * X_new + 2, color='red', linewidth=2, linestyle='--', label='True line: y=3x+2')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression: Fitted vs True Line')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Calculate R score
from sklearn.metrics import r2_score, mean_squared_error
y_pred = model.predict(X_new)
r2 = r2_score(y_new, y_pred)
mse = mean_squared_error(y_new, y_pred)

print(f"\nModel Performance Metrics:")
print(f"R Score: {r2:.4f}")
print(f"Mean Squared Error: {mse:.4f}")
```
