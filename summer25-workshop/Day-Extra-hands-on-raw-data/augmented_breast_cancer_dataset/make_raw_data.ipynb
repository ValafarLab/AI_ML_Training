```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
import random
import numpy as np
```

```python
obj = load_breast_cancer()
```

```python
print(obj['DESCR'])
```

```python
X = obj['data']
y = obj['target']
X.shape, y.shape
```

```python
print([type(x) for x in X[0]])

print('\n')
print(obj['feature_names'],'\n')
print(obj['target_names'])
```

```python
df = pd.DataFrame(data=X, columns=obj['feature_names'])
```

```python
df['diagnosis'] = y
```

```python
df['diagnosis'] = df['diagnosis'].apply(lambda x: 'benign' if x == 0 else 'malignant')
```

```python
random.sample(list(np.arange(1000, 5000, 1)), 569)
```

```python
df['id'] = random.sample(list(np.arange(1000, 5000, 1)), 569)
df
```

```python
colls = list(df.columns)
colls.remove('id')
colls.insert(0,'id')

df = df[colls]
df
```

```python
# Now we start to modify the df an mess it up

# 1. add rows with missing data


#duplicate some rows then we will cahnge the id 

duprows = df.iloc[random.choices(np.arange(start=0, stop=568, step=1), k=57)]

# drop between 1 - 7 of the colons in each row to nan
for idx, row in duprows.iterrows():
    num_colls_to_drop = random.choice(range(1,8)) # how many to drop betwen 1 - 7
    coll_idxs_to_drop =  random.sample(range(1,32),k=num_colls_to_drop)#which of the 31 colls shoudld we drop
    duprows.loc[idx, duprows.columns[coll_idxs_to_drop]] = np.nan


#chacnge the id's to be the new in ragne from 5000 - 7000
new_ids = random.sample(range(5000, 7001), k=len(duprows))
duprows.iloc[:, 0] = new_ids 

duprows
```

```python
df_messed_up = pd.concat([df, duprows], axis=0)
df_messed_up
```

```python
# 2. lets make some colls from floats to strings

# lets pick 5 random colls
colls_to_switch_type = random.sample(range(1,33), k=5)
print(colls_to_switch_type)
df_messed_up.iloc[:, colls_to_switch_type] = df_messed_up.iloc[:, colls_to_switch_type].astype('str')

df_messed_up.iloc[:,colls_to_switch_type] = df_messed_up.iloc[:,colls_to_switch_type].astype('str')
df_messed_up.describe()
```

```python
#3 last thing we will do is duplicate some rows

# 15 idxs to duplicate
rows_to_dup = random.choices(range(0,626), k=15)
dup_rows = df_messed_up.iloc[rows_to_dup]
df_messed_up = pd.concat([df_messed_up, dup_rows], axis=0, ignore_index=True)
df_messed_up
```
